## 📅 Mois 1 – Analyse Marketing basée sur les données de ventes (2014–2017)

Ce projet marque le **début de mon défi "Devenir Data Analyst en 9 mois"**.

Pour ce premier mois, j’ai travaillé sur un jeu de données de ventes réelles enregistrées entre **2014 et 2017** par un géant du retail, dans un contexte de **concurrence intense et de recherche de performance**.  
L'entreprise cherche à comprendre **quels produits, régions, catégories et segments clients** cibler ou éviter pour optimiser ses résultats.

J’ai mené une **analyse marketing 100 % data-driven**, en combinant :
- **analyses univariées** pour comprendre la distribution de chaque variable,
- **analyses bivariées** pour détecter les relations clés entre variables,
- et un **tableau de bord interactif dans Excel** pour visualiser dynamiquement les performances par segment, région, produit, etc.

Mon analyse a mis en lumière des leviers d’action stratégiques :

📌 Quels produits génèrent réellement du profit?  

📌 Quelles réductions nuisent à la rentabilité? 

📌 Quels clients et régions méritent d’être mieux ciblés? 

Grâce à ces insights, **l’entreprise peut affiner ses décisions marketing et commerciales pour maximiser ses profits tout en réduisant les pertes**.

## 📅 Month 1 – Marketing Analysis Based on Sales Data (2014–2017)

This project marks the **start of my "Becoming a Data Analyst in 9 Months" challenge**.

For this first month, I worked on a real sales dataset recorded between **2014 and 2017** by a superstore giant, in a context of **intense competition and performance-driven goals**.  
The company seeks to understand **which products, regions, categories, and customer segments** to target or avoid in order to optimize results.

I conducted a **100% data-driven marketing analysis**, combining:  
- **univariate analyses** to understand the distribution of each variable,  
- **bivariate analyses** to detect key relationships between variables,  
- and an **interactive dashboard in Excel** to dynamically visualize performance by segment, region, product, etc.

My analysis highlighted strategic action levers:  
📌 Which products truly generate profit?  

📌 Which discounts harm profitability?  

📌 Which customers and regions deserve better targeting?

Thanks to these insights, **the company can refine its marketing and commercial decisions to maximize profits while reducing losses**.
